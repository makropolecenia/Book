```python
https://covidtrackerapi.bsg.ox.ac.uk/api/v2/stringency/date-range/{YYYY-MM-DD}/{YYYY-MM-DD}
where last two values represent start and end dates
```

```python
https://covidtrackerapi.bsg.ox.ac.uk/api/v2/stringency/actions/{ALPHA-3}/{YYYY-MM-DD}
where last two values represent an ISO 3166-1 alpha-3 country code and a date 
```

```python
#https://stackoverflow.com/questions/32400867/pandas-read-csv-from-url
import pandas as pd
import io
import requests

url = 'https://covidtrackerapi.bsg.ox.ac.uk/api/v2/stringency/date-range/2021-02-01/2021-04-07'
s=requests.get(url).content
c=pd.read_csv(io.StringIO(s.decode('utf-8')))

print(c)
c
```

```python
#https://stackoverflow.com/questions/32400867/pandas-read-csv-from-url
import pandas as pd
import io
import requests

url = 'https://covidtrackerapi.bsg.ox.ac.uk/api/v2/stringency/actions/122/2021-05-27'

s=requests.get(url).content
c=pd.read_csv(io.StringIO(s.decode('utf-8')))

print(c)
c

```
